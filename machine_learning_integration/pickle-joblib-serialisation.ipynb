```python
import pickle
import joblib
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('D:/learn/fast-api/machine_learning_integration/housing.csv').iloc[:, :-1].dropna()
df
```

```python
x = df.drop(columns='median_house_value')
y = df.median_house_value.copy()

print(x.shape, y.shape)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
r2_score(y, y_pred)
```

```python
model
```

```python
with open('linear_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
with open('linear_model.pkl', 'rb') as f:
    pickle_model = pickle.load(f)
```

```python
pickle_model
```

```python
y_pred_pickle = pickle_model.predict(x)
y_pred_pickle
```

```python
joblib.dump(model, 'linear_model_joblib.joblib')
```

```python
joblib_model = joblib.load('linear_model_joblib.joblib')
joblib_model
```

```python
y_pred_joblib = joblib_model.predict(x)
y_pred_joblib
```

```python
r2_score(y, y_pred_joblib)
```
